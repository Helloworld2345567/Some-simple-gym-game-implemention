```python
import numpy as np
import matplotlib.pyplot as plt

rewards = np.load('./logs/rewards_log0.npy')
average = [np.mean(rewards[i-100:i]) for i in range(100, len(rewards))]

plt.plot(rewards,alpha=0.5, label='rewards')
plt.plot(average,'b', label='average')
plt.xlabel('Ep')
plt.ylabel('rewards')
plt.legend()
plt.show()
```
